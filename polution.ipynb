```python
app = Dash(__name__)

app.layout = html.Div([
    html.H1(children='Title of Dash App', style={'textAlign':'center'}),
    dash_table.DataTable(data=polution_dataset.to_dict('records'), page_size=10),
    dcc.Dropdown(polution_dataset.Country.unique(), ' ', id='dropdown-selection'),
    dcc.Graph(id='graph-content')
])

@callback(
    Output('graph-content', 'figure'),
    Input('dropdown-selection', 'value')
)
def update_graph(value):
    dff = polution_dataset[polution_dataset.Country==value]
    return px.line(dff, x='Year', y='TotalDeaths')

if __name__ == '__main__':
    app.run(debug=True)
```
